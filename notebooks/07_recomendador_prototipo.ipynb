PROTOTIPO RECOMENDADOR MOTOFIT

Bienvenido!
Elegir una moto nueva entre tantas alternativas puede ser abrumador, pero algunas simples preguntasnos pueden ayudar.

¬øCu√°l es tu presupuesto m√°ximo? - ¬øQu√© carnet tienes (A1, A2, A)? - ¬øCu√°l es tu estatura en cm?

DATO: Dado que solo tendremos la altura del piloto y no podremos obtener la medida exacta de la entrepierna, utilizaremos una f√≥rmula de estimaci√≥n, com√∫n en la ergonom√≠a, para obtener la altura del asiento que consideraremos como m√°s adecuada al tama√±o del usuario. Destacando que es una aproximaci√≥n para guiar la b√∫squeda y no una medida absoluta, puesto que entre otros factores depender√° tambi√©n del anvcho de la moto, del tipo de uso y la experiencia del piloto, entre otros factores. 

```python
# Asegurar tipos num√©ricos correctos
for col in ['PRECIO', 'ALTURA_ASIENTO']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Funci√≥n de recomendaci√≥n
def recomendar_motos(presupuesto, carnet, altura_usuario):
    df_filtrado = df.copy()

    # Filtro por precio
    df_filtrado = df_filtrado[df_filtrado['PRECIO'] <= presupuesto]

    # Filtro por carnet
    carnet_orden = {'A1': 1, 'A2': 2, 'A': 3}
    carnet_usuario = carnet_orden[carnet.upper()]
    df_filtrado = df_filtrado[df_filtrado['CARNET_MINIMO'].map(carnet_orden) <= carnet_usuario]

    # Filtro por altura (en mm)
    altura_max = altura_usuario * 4.6 - 30
    print(f"\nüîé Altura m√°xima de asiento sugerida: {altura_max:.0f} mm\n")
    df_filtrado = df_filtrado[df_filtrado['ALTURA_ASIENTO'] <= altura_max]

    # Ordenar por precio
    df_resultado = df_filtrado.sort_values(by='PRECIO')

    # Mostrar columnas clave
    columnas_mostrar = ['MARCA', 'MODELO', 'TIPO_SIMPLIFICADO', 'PRECIO', 'ALTURA_ASIENTO', 'CARNET_MINIMO', 'PESO_VACIO', 'POTENCIA']
    return df_resultado[columnas_mostrar].reset_index(drop=True)

# Inputs desde consola
print("üõµ Bienvenido al recomendador MotoFit üß†\n")
presupuesto = float(input("üí∞ ¬øCu√°l es tu presupuesto m√°ximo en euros? "))
carnet = input("üé´ ¬øQu√© carnet tienes (A1, A2, A)? ")
altura = int(input("üìè ¬øCu√°l es tu estatura (en cm)?"))

# Ejecutar recomendaci√≥n
resultado = recomendar_motos(presupuesto, carnet, altura)

# Mostrar resultados
if resultado.empty:
    print("\n‚ùå Lo sentimos, no encontramos ninguna moto que cumpla con esos criterios.\n")
else:
    print("\n‚úÖ Motos recomendadas para ti:\n")
    print(resultado.to_string(index=False))

```

PROTOTIPO MEJORADO DEL GRAFICO QUE MUESTRA POTENCIA VS DINERO GASTADO USANDO IA. EL anterior (notebook 05) es soso y aburrido, necesito mejorarlo para compartir.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker # Para formatear los ejes
import seaborn as sns # Seaborn es excelente para gr√°ficos m√°s bonitos

# Asumo que df ya est√° cargado y limpio de la ejecuci√≥n anterior
# 1. Cargar los datos (si no lo has hecho ya en el mismo script)

df = pd.read_csv("motofit_limpio.csv")

# 2. Asegurar que POTENCIA y PRECIO sean num√©ricos (lo tienes bien)
df['POTENCIA'] = pd.to_numeric(df['POTENCIA'], errors='coerce')
df['PRECIO'] = pd.to_numeric(df['PRECIO'], errors='coerce')

# 3. Eliminar las motos sin datos de potencia o precio (lo tienes bien)
df_clean = df.dropna(subset=['POTENCIA', 'PRECIO'])

# 4. Calcular cu√°ntos caballos entrega cada moto por cada 100‚Ç¨ (lo tienes bien)
df_clean['POTENCIA_POR_100_EUROS'] = (df_clean['POTENCIA'] / df_clean['PRECIO']) * 100

# 5. Agrupar por tipo de moto y calcular el promedio de esa m√©trica
# Importante: para barh, ordenamos ascending=True para que el mayor quede arriba al graficar
potencia_por_100 = df_clean.groupby('TIPO_SIMPLIFICADO')['POTENCIA_POR_100_EUROS'].mean().sort_values(ascending=True)

# --- 6. Crear el gr√°fico con mejoras est√©ticas ---
plt.style.use('seaborn-v0_8-darkgrid') # Un estilo de seaborn que mejora el aspecto general
# Tambi√©n puedes probar 'ggplot', 'seaborn-v0_8-pastel', etc.

plt.figure(figsize=(12, 7)) # Aumentamos el tama√±o para mejor visibilidad

# Crear el gr√°fico de barras HORIZONTALES
bars = plt.barh(potencia_por_100.index, potencia_por_100.values, color=sns.color_palette("viridis", len(potencia_por_100)))

plt.title('üêé CV Promedio por cada 100‚Ç¨ seg√∫n Tipo de Moto (¬°Eficiencia en Acci√≥n!)',
          fontsize=16, pad=20, color='#333333') # T√≠tulo m√°s grande y con color
plt.xlabel('CV por cada 100‚Ç¨ (Promedio)', fontsize=12, labelpad=15, color='#555555') # Etiqueta del eje X
plt.ylabel('Tipo de Moto', fontsize=12, labelpad=15, color='#555555') # Etiqueta del eje Y

# A√±adir los valores num√©ricos exactos al final de cada barra
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2,
             f'{bar.get_width():.2f}', # Formatear a 2 decimales
             va='center', ha='left', fontsize=10, color='black')

# Mejorar el dise√±o de los ejes y rejillas
plt.xticks(fontsize=10, color='#666666')
plt.yticks(fontsize=11, color='#666666')
plt.gca().invert_yaxis() # Para asegurar que la barra m√°s larga (mayor valor) est√© en la parte superior

plt.grid(axis='x', linestyle='--', alpha=0.5) # Rejillas horizontales m√°s sutiles
plt.grid(axis='y', linestyle='', alpha=0) # Eliminar rejillas verticales

# Eliminar los "spines" (bordes) para una apariencia m√°s limpia
sns.despine(left=True, bottom=True) # Elimina los bordes del gr√°fico

plt.tight_layout() # Ajusta el dise√±o para que no se superpongan elementos
plt.show()
```
