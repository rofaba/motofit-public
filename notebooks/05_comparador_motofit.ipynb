¬øCon qu√© se encuentra alguien que quiere elegir una moto?

- Una inmensa variedad: Miles de modelos con nombres y caracter√≠sticas similares pero diferencias sutiles.

- Jergas t√©cnicas: Mucha terminolog√≠a (V-Twin, par motor, DCT, etc.) que puede abrumar.

- Necesidades vs. Deseos: A menudo, lo que creen que quieren no se ajusta a lo que necesitan realmente para su uso diario o experiencia.

- Presupuesto: El factor limitante n√∫mero uno.

- Percepci√≥n de marca: La fama, el prestigio o la reputaci√≥n de una marca.

- Factores emocionales: El dise√±o, el sonido, la "sensaci√≥n" al subirse.

Qu√© podemos considerar y utilizar, de cara a un comparador o recomendador.

1. Datos Esenciales
- MARCA: Crucial para la fidelidad a la marca o para buscar alternativas.
- MODELO: Identificador √∫nico de cada moto.
- TIPO DE MOTO: Quiero una naked, una trail, una scooter...

- Cilindrada (cc): Importante. Define directamente la potencia, el tipo de carnet necesario (A1, A2, A) y el segmento de la moto.
¬øPor qu√© es importante?: Restricciones de carnet, tipo de uso (urbano, carretera, viajes), coste del seguro, consumo.

- Potencia (CV o kW): Ligado a la cilindrada, pero no lineal.
¬øPor qu√© es importante?: Sensaciones de conducci√≥n, aceleraci√≥n, velocidad m√°xima, aptitud para viajar con pasajero o carga.

- Tipo de Motor: (Ej: Monocil√≠ndrico, Bicil√≠ndrico en l√≠nea, V-Twin, Cuatro cilindros en l√≠nea, B√≥xer). 
Cada configuraci√≥n tiene un sonido, una entrega de potencia y unas vibraciones caracter√≠sticas.
ES UN DATO QUE NO TENEMOS DISPONIBLE POR AHORA

- Tipo de Transmisi√≥n: (Manual, DCT, Autom√°tica para scooters).
¬øPor qu√© es importante?: Facilidad de conducci√≥n (especialmente DCT/autom√°tico en ciudad), experiencia de conducci√≥n.
NO LO TENEMOS DIRECTAMENTE, EN ALGUNOS CASOS SE PUEDE INFERIR DESDE EL TIPO

- Peso (kg): Influye enormemente en la manejabilidad, especialmente en parado o a baja velocidad.
¬øPor qu√© es importante?: Facilidad para principiantes, manejo en ciudad, capacidad de maniobra, facilidad de aparcamiento.

- Altura del Asiento (mm): Dato ergon√≥mico vital para la comodidad y seguridad del piloto.
¬øPor qu√© es importante?: Permite al piloto llegar al suelo con los pies, lo que da confianza, especialmente a personas de menor estatura o principiantes.
- Capacidad del Dep√≥sito (Litros) y Autonom√≠a (km): Para viajes largos.
Consumo de Combustible (L/100km): Afecta directamente el coste de uso.
NO DISPONIBLE AUN


3. Datos de Equipamiento y Tecnolog√≠a (a√∫n no incoporados en esta etapa de desarrollo):
Sistema de Frenado (ABS, CBS): La seguridad es primordial. El ABS es casi un est√°ndar hoy en d√≠a.
Electr√≥nica Avanzada: (Control de tracci√≥n, modos de conducci√≥n, control de crucero, quickshifter, etc.). Estos sistemas mejoran la seguridad y la experiencia de conducci√≥n.
Tipo Iluminaci√≥n: (LED, hal√≥gena).
Panel de Instrumentos: (Digital, TFT, anal√≥gico).
Conectividad: (Bluetooth, GPS integrado).
Equipamiento Espec√≠fico: (Maletas integradas, defensas, c√∫pulas ajustables, pu√±os calefactables, etc.). Esto es crucial para motos touring o adventure.

4. Datos de Precio y Mercado:

- Precio Nuevo (EUR): Absolutamente fundamental. 
El presupuesto es el primer filtro para la mayor√≠a de los compradores.

PENDIENTES
- Precio de Reventa / Valor Usado (EUR): Aunque m√°s complejo de obtener, es un dato que muchos compradores consideran impl√≠citamente.

- Disponibilidad (Nuevo/Usado): ¬øSe sigue fabricando? ¬øEs un modelo que solo se encuentra de segunda mano?


5. Datos de la Experiencia del Usuario (M√°s avanzados para un recomendador, pero valiosos para el an√°lisis):
- Puntuaciones/Valoraciones de Usuarios: Opiniones subjetivas de quienes ya poseen la moto.


- Comentarios y Rese√±as: Informaci√≥n cualitativa sobre problemas comunes, virtudes inesperadas, etc.

- Fiabilidad General: Percepci√≥n de la marca/modelo en cuanto a aver√≠as.

```python
# He probado con diferentes caracter√≠sticas, el dataset se ve bastante correcto.

# Cargar dataset
df = pd.read_csv('motofit_v3.csv')

# Asegurar num√©rico
df['POTENCIA'] = pd.to_numeric(df['POTENCIA'], errors='coerce')

# Calcular IQR para la columna
Q1 = df['POTENCIA'].quantile(0.25)
Q3 = df['POTENCIA'].quantile(0.75)
IQR = Q3 - Q1

# Definir l√≠mites superior e inferior
limite_bajo = Q1 - 1.5 * IQR
limite_alto = Q3 + 1.5 * IQR

# Filtrar outliers
outliers_peso = df[(df['POTENCIA'] < limite_bajo) | (df['POTENCIA'] > limite_alto)]

# Mostrar posibles errores
print(outliers_peso[['MARCA', 'MODELO', 'POTENCIA', 'TIPO_SIMPLIFICADO']])
```
