¿Con qué se encuentra alguien que quiere elegir una moto?

- Una inmensa variedad: Miles de modelos con nombres y características similares pero diferencias sutiles.

- Jergas técnicas: Mucha terminología (V-Twin, par motor, DCT, etc.) que puede abrumar.

- Necesidades vs. Deseos: A menudo, lo que creen que quieren no se ajusta a lo que necesitan realmente para su uso diario o experiencia.

- Presupuesto: El factor limitante número uno.

- Percepción de marca: La fama, el prestigio o la reputación de una marca.

- Factores emocionales: El diseño, el sonido, la "sensación" al subirse.

Qué podemos considerar y utilizar, de cara a un comparador o recomendador.

1. Datos Esenciales
- MARCA: Crucial para la fidelidad a la marca o para buscar alternativas.
- MODELO: Identificador único de cada moto.
- TIPO DE MOTO: Quiero una naked, una trail, una scooter...

- Cilindrada (cc): Importante. Define directamente la potencia, el tipo de carnet necesario (A1, A2, A) y el segmento de la moto.
¿Por qué es importante?: Restricciones de carnet, tipo de uso (urbano, carretera, viajes), coste del seguro, consumo.

- Potencia (CV o kW): Ligado a la cilindrada, pero no lineal.
¿Por qué es importante?: Sensaciones de conducción, aceleración, velocidad máxima, aptitud para viajar con pasajero o carga.

- Tipo de Motor: (Ej: Monocilíndrico, Bicilíndrico en línea, V-Twin, Cuatro cilindros en línea, Bóxer). 
Cada configuración tiene un sonido, una entrega de potencia y unas vibraciones características.
ES UN DATO QUE NO TENEMOS DISPONIBLE POR AHORA

- Tipo de Transmisión: (Manual, DCT, Automática para scooters).
¿Por qué es importante?: Facilidad de conducción (especialmente DCT/automático en ciudad), experiencia de conducción.
NO LO TENEMOS DIRECTAMENTE, EN ALGUNOS CASOS SE PUEDE INFERIR DESDE EL TIPO

- Peso (kg): Influye enormemente en la manejabilidad, especialmente en parado o a baja velocidad.
¿Por qué es importante?: Facilidad para principiantes, manejo en ciudad, capacidad de maniobra, facilidad de aparcamiento.

- Altura del Asiento (mm): Dato ergonómico vital para la comodidad y seguridad del piloto.
¿Por qué es importante?: Permite al piloto llegar al suelo con los pies, lo que da confianza, especialmente a personas de menor estatura o principiantes.
- Capacidad del Depósito (Litros) y Autonomía (km): Para viajes largos.
Consumo de Combustible (L/100km): Afecta directamente el coste de uso.
NO DISPONIBLE AUN


3. Datos de Equipamiento y Tecnología (aún no incoporados en esta etapa de desarrollo):
Sistema de Frenado (ABS, CBS): La seguridad es primordial. El ABS es casi un estándar hoy en día.
Electrónica Avanzada: (Control de tracción, modos de conducción, control de crucero, quickshifter, etc.). Estos sistemas mejoran la seguridad y la experiencia de conducción.
Tipo Iluminación: (LED, halógena).
Panel de Instrumentos: (Digital, TFT, analógico).
Conectividad: (Bluetooth, GPS integrado).
Equipamiento Específico: (Maletas integradas, defensas, cúpulas ajustables, puños calefactables, etc.). Esto es crucial para motos touring o adventure.

4. Datos de Precio y Mercado:

- Precio Nuevo (EUR): Absolutamente fundamental. 
El presupuesto es el primer filtro para la mayoría de los compradores.

PENDIENTES
- Precio de Reventa / Valor Usado (EUR): Aunque más complejo de obtener, es un dato que muchos compradores consideran implícitamente.

- Disponibilidad (Nuevo/Usado): ¿Se sigue fabricando? ¿Es un modelo que solo se encuentra de segunda mano?


5. Datos de la Experiencia del Usuario (Más avanzados para un recomendador, pero valiosos para el análisis):
- Puntuaciones/Valoraciones de Usuarios: Opiniones subjetivas de quienes ya poseen la moto.


- Comentarios y Reseñas: Información cualitativa sobre problemas comunes, virtudes inesperadas, etc.

- Fiabilidad General: Percepción de la marca/modelo en cuanto a averías.
