```python
import pandas as pd
import os

# Ruta donde están todos los CSV
ruta_csv = "../data/raw/"
archivos_csv = [f for f in os.listdir(ruta_csv) if f.endswith(".csv")]

# Diccionario de renombrado de columnas estandarizadas
renombrar_columnas = {
    "modelo": "MODELO",
    "Modelo": "MODELO",
    "MODELO": "MODELO",

    "carnet": "CON_CARNET",
    "CARNET": "CON_CARNET",
    "Con_Carnet": "CON_CARNET",
    "CON_CARNET": "CON_CARNET",

    "cilindrada": "CILINDRADA",
    "CILINDRADA": "CILINDRADA",
    "cilindrada_cm3": "CILINDRADA",
    "cc": "CILINDRADA",

    "potencia": "POTENCIA",
    "Potencia": "POTENCIA",
    "potencia_cv": "POTENCIA",
    "POTENCIA": "POTENCIA",
    "POTENCIA_CV": "POTENCIA",

    "rueda_delantera": "RUEDA_DELANTERA",
    "neumatico_delantero": "RUEDA_DELANTERA",
    "NEUMATICO_DELANTERO": "RUEDA_DELANTERA",
    "NEUMÁTICO_DELANTERO": "RUEDA_DELANTERA",
    "Rueda_Delantera": "RUEDA_DELANTERA",
    "Rueda Delantera": "RUEDA_DELANTERA",

    "rueda_trasera": "RUEDA_TRASERA",
    "neumatico_trasero": "RUEDA_TRASERA",
    "NEUMATICO_TRASERO": "RUEDA_TRASERA",
    "NEUMÁTICO_TRASERO": "RUEDA_TRASERA",
    "Rueda_Trasera": "RUEDA_TRASERA",
    "Rueda Trasera": "RUEDA_TRASERA",

    "freno_delantero": "FRENO_DELANTERO",
    "FRENO_DELANTERO": "FRENO_DELANTERO",

    "freno_trasero": "FRENO_TRASERO",
    "FRENO_TRASERO": "FRENO_TRASERO",

    "altura_asiento": "ALTURA_ASIENTO",
    "ALTURA_ASIENTO": "ALTURA_ASIENTO",

    "peso_vacio": "PESO_VACIO",
    "peso": "PESO_VACIO",
    "PESO_VACIO": "PESO_VACIO",

    "precio": "PRECIO",
    "PRECIO": "PRECIO"
}

# Lista para guardar DataFrames
lista_df = []

for archivo in archivos_csv:
    df = pd.read_csv(os.path.join(ruta_csv, archivo))
    
    # Renombrar columnas estandarizadas
    df = df.rename(columns=lambda x: renombrar_columnas.get(x.strip(), x.strip()))
    
    # Agregar columna de marca
    marca = archivo.replace(".csv", "").upper()
    df["MARCA"] = marca
    
    lista_df.append(df)

# Concatenar todos los DataFrames
df_consolidado = pd.concat(lista_df, ignore_index=True)

# Guardar el resultado
df_consolidado.to_csv("../data/dataset_motos_consolidado.csv", index=False)


```

ahora necesitamos determinar el carnet minimo necesario para conducir cada moto para poder categorizarlas.

```python

import numpy as np

df_consolidado = pd.read_csv("motofit_dataset_limpio.csv")

orden_carnets = {"AM": 1, "A1": 2, "A2": 3, "A": 4, "B": 5}
inverso_carnets = {v: k for k, v in orden_carnets.items()}

def carnet_minimo(c):
    if pd.isna(c):
        return np.nan
    niveles = [orden_carnets[p] for p in c.replace("/", "-").replace(" ", "").split("-") if p in orden_carnets]
    return inverso_carnets[min(niveles)] if niveles else np.nan

df_consolidado["CARNET_MINIMO"] = df_consolidado["CON_CARNET"].apply(carnet_minimo)

```

```python
# mediante una fn aplicamos palabras clave para intentar categorizar 
# las motos por tipo. Iteración 4.

def clasificar_tipo_moto_mejorado_v4(modelo):
    """
    Clasifica el tipo de motocicleta basado en el nombre del modelo.
    Se han añadido y ampliado palabras clave para una mejor detección
    y se ha incluido la categoría 'Custom' de forma más robusta.

    Args:
        modelo (str): El nombre del modelo de la motocicleta.

    Returns:
        str: El tipo de motocicleta clasificado (ej: "Adventure", "Naked", "Scooter", etc.)
             o "Otro" si no se puede clasificar.
    """
    if pd.isna(modelo):
        return "Otro"
    m = str(modelo).lower()

    # Adventure (Aventura/Trail): Motos diseñadas para viajes largos y terrenos variados.
    # Incluye modelos con 'adv', 'gs', 'trail', 'tenere', 'africa twin', 'multistrada', etc.
    if any(k in m for k in ["adv", "adventure", "gs", "trv", "norden", "tenere", "desert", "rally", "trail", "v-strom", "dsx", "trk", "stelvio",
                            "transalp", "caponord", "versys", "tiguer", "desertx", "desert x", "multistrada", "himalayan", "cbf190x", "pan america",
                            "guerrilla", "tuareg", "800mt", "nx500"]):
        return "Adventure"
    # Cruiser: Motos con postura relajada, asientos bajos, a menudo con motores grandes.
    # Incluye 'bobber', 'rebel', 'intruder', 'scout', 'road glide', 'fat boy', 'electra glide', etc.
    elif any(k in m for k in ["bobber", "cruiser", "rebel", "intruder", "sportster", "diavel", "scout", "road glide", "fat bob", "heritage",
                              "softail", "nightster", "electra glide", "fat boy", "iron", "forty-eight", "street bob", "shadow", "vulcan",
                              "v-cruise", "eliminator", "road king", "freewheeler", "breakout", "springfield", "chief"]):
        return "Cruiser"
    # Scrambler: Motos de estilo retro, a menudo con neumáticos mixtos y escape elevado.
    # Incluye 'scrambler', 'icon dark', 'desert sled', 'urban enduro', 'fantic caballero', 'cl500'.
    elif any(k in m for k in ["scrambler", "icon dark", "desert sled", "urban enduro", "nigthshift", "fantic cabal", "cl500", "scram",
                              "street scrambler", "scrambler 400", "scrambler 800", "scrambler 1100", "xcr", "crossfire", "caballero"]):
        return "Scrambler"
    # Sport (Deportiva): Motos diseñadas para velocidad y rendimiento en pista.
    # Incluye 'panigale', 'ninja', 'cbr', 'zx', 'r', 'gsx', 'supersport', 'hayabusa', etc.
    elif any(k in m for k in ["panigale", "ninja", "cbr", "zx", " rr", "rc", "gsx", "supersport", "hayabusa", "gsxr", "yzf", "s1000rr", "zx10r", "r6", "r1",
                              "aprillia rs", "daytona", "fz", "gixxer", "rsv4", "rc390", "yzf-r", "srk", "450sr", "703rr", "zx-6r", "zx-10r", "fireblade",
                              "gsx-r", "h2"]):
        return "Sport"
    # Naked: Motos sin carenado, mostrando el motor y chasis, con postura de conducción más erguida.
    # Incluye 'mt', 'duke', 'sv', 'cb', 'hornet', 'z', 'monster', 'leoncino', 'katana', etc.
    elif any(k in m for k in ["mt", "duke", "sv ", "cb", "hornet", "z ", "z900", "z650", "monster", "vitpilen", "svartpilen", "fz", "fz6", "street triple", "brutale",
                              "368g", "350d", "gk350s", "streetfighter", "cb650r", "bandit", "griso", "tuono", "rayzr", "cb1000r", "xjr",
                              "street twin", "speed twin", "bonneville", "w800", "retro", "leoncino", "imperiale", "katana", "nk", "s1000r", "m1000r",
                              "752s", "srk", "502n", "808", "v302c", "tnt", "bn", "rkf", "x-light", "mbp m502n", "r nine t", "rninet", "r12", "hunter", "bullet", "classic 350", "meteor", "g310r", "clx", "300cl-x", "450nk", "650nk", "nks", "rkr", "z500", "v7", "v9", "v85", "v100", "pm-03", "sv650", "trident", "speed 400", "125r", "350ac", "525r", "hypermotard", "ftr", "xsr"]):
        return "Naked"
    # Tourer (Touring/Gran Turismo): Motos diseñadas para viajes largos con comodidad y capacidad de carga.
    # Incluye 'tour', 'gt', 'goldwing', 'tracer', 'k1600', 'rt', 'nt1100', 'gsx-s1000gt', etc.
    elif any(k in m for k in ["tour", "gt", "goldwing", "nt1100", "c 400", "c400", "burgman", "maxsym", "forza", "tracer", "tourer", "touring", "fjr", "k1600",
                              "rt", "glt", "k 1600", "rs ", "transcontinental", "grand america", "gsx-s1000gt", "versys 1000", "street glide",
                              "tri glide ultra", "rocket 3", "rocket3", "metropolis", "pulsion", "beverly", "xciting", "ak 550"]):
        return "Tourer"
    # Off-road (Todoterreno/Enduro/Motocross): Motos para uso fuera de carretera.
    # Incluye 'enduro', 'xr', 'crf', 'drz', 'wr', 'klx', 'dirt', 'motocross', 'sx', 'exc', 'rr'.
    elif any(k in m for k in ["enduro", "xr", "crf", "drz", "wr", "klx", "dirt", "xtrainer", "cross", "300 rally", "motocross", "offroad",
                              "sx", "exc", "fe", "te", "tx", "fs", "rm-z", "kx", "yz", "rx ", "rr ", "beta", "montana", "supermoto", "smc r", "450 smr", "g450x"]):
        return "Off-road"
    # Scooter: Vehículos de dos ruedas con plataforma para los pies y a menudo con carrocería protectora.
    # Incluye 'scooter', 'pcx', 'xmax', 'nmax', 'dtx', 'tmax', 'vespa', 'joyride', 'jet x', 'mp3', 'primavera', 'tricity'.
    elif any(k in m for k in ["scooter", "pcx", "xmax", "nmax", "dtx", "tmax", "sr ", "vespa", "symphony", "people", "joyride", "jet x", "vieste", "mp3", "primavera",
                              "medley", "agility", "super dink", "x-town", "adv350", "sh", " forza", "neos", "d'elight", "tricity", "niu", "nqi", "mqi", "ngt",
                              "em1 e", "vision", "tweet", "kisbee", "speedfight", "django", "zip", "piaggio 1", "address", "avenis", "fiddle", "adx", "joymax", "cruisym"]):
        return "Scooter"
    # Custom: Motos personalizadas o con un estilo muy particular, a menudo derivado de las cruisers pero con más énfasis en la estética única.
    # Incluye 'custom', 'chopper', 'lowrider', 'nightster', 'bagger', 'road king', 'street glide'.
    elif any(k in m for k in ["custom", "chopper", "lowrider", "nightster", "r 18", "breakout", "softail slim", "fat bob", "cvo", "livewire", "s2", "super cub", "monkey", "dax"]):
        return "Custom"
    else:
        return "Otro"


```

Many of the TIPO_CORREGIDO classifications are accurate. However, I found several instances where the assigned type could be more precise or is generally considered incorrect based on common industry classification and online information.

Scooters Misclassified as "Tourer", "Off-road", or "Adventure": Many maxi-scooters or urban scooters were labeled as "Tourer" (e.g., Aprilia SR GT, BMW C 400, Honda Forza, Kymco Xciting/AK, MITT RT/XRS 50, Peugeot Pulsion/Metropolis, Piaggio Liberty/Beverly, SYM Maxsym, Vespa GTS 310) or "Adventure" (e.g., Honda ADV350, Honda X-ADV, Peugeot XP400). Cruisers/Standards Misclassified as "Naked", "Off-road", or "Custom": Some cruisers (e.g., Benelli Imperiale 400, Royal Enfield Meteor 350, Royal Enfield Super Meteor 650, Triumph Bonneville Speedmaster) were labeled as "Naked" or "Scrambler". Sport Tourers Misclassified as "Sport" or "Naked": Performance-oriented touring bikes (e.g., BMW S 1000 XR, BMW R 1250 RS, Kawasaki Ninja 1000 SX, Kawasaki Ninja H2 SX, KTM 1390 Super Duke GT, Suzuki GSX-S1000GT) were often simply "Sport" or "Naked" instead of "Sport Tourer".
